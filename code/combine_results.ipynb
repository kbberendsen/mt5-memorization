# Duplicates

```python
titles = ['ad', 'fd', 'groene', 'nrc', 'volkskrant']
mask_levels = ['15', '20', '25', '30']
model_sizes = ['small', 'base', 'large', 'xl', 'xxl']

RESULTS_PATH = '..\\results\\working_results\\mT5\\duplicates'
DATA_PATH = '..\\data\\duplicates\\masked_input'
```

```python
# Concat results of all titles for a single masking level
df_results_15, df_results_20, df_results_25, df_results_30 =  pd.DataFrame(), pd.DataFrame(), pd.DataFrame(), pd.DataFrame()

for mask_level in mask_levels:
    df_results_title = pd.DataFrame()
    for title in titles:
        df_results = pd.DataFrame()
        for model_size in model_sizes:          
            try:
                df_results_temp = pd.read_csv(f'{RESULTS_PATH}\\results_{mask_level}\\{title}\\results_{title}_{model_size}.csv')
                df_results_temp['Title'] = title
                df_results = pd.concat([df_results, df_results_temp], ignore_index=True)
            except Exception as e:
                print(f'Results not available for {title}, {mask_level}% mask, {model_size} model: {e}')
                continue

        try: 
            df_results_title = pd.concat([df_results_title, df_results], ignore_index=True)
        except:
            continue        

    try: 
        if mask_level == '15':
            df_results_15 = pd.concat([df_results_15, df_results_title], ignore_index=True)
        elif mask_level == '20':
            df_results_20 = pd.concat([df_results_20, df_results_title], ignore_index=True)
        elif mask_level == '25':
            df_results_25 = pd.concat([df_results_25, df_results_title], ignore_index=True)
        elif mask_level == '30':
            df_results_30 = pd.concat([df_results_30, df_results_title], ignore_index=True)
    except:
        continue
```
