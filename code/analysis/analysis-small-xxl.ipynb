```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
# Read the CSV file
df = pd.read_csv(r"..\results\20240507\test_results_groene_merged.csv", delimiter=';')
df.head()
```

```python
# Extract data
lev_small = df['lev_small']
lev_base = df['lev_base']
lev_large = df['lev_large']
lev_xl = df['lev_xl']
lev_xxl = df['lev_xxl']

# Calculate the bin edges manually
bins = 5
bin_edges = range(min(min(lev_base), min(lev_large)), max(max(lev_base), max(lev_large)) + bins, bins)

# Create a figure and axis objects
fig, ax = plt.subplots(figsize=(10, 6))

# Plot histogram for lev_small
ax.hist(lev_small, bins=bin_edges, color='yellow', alpha=0.9, label='lev_small', align='left')
# Plot histogram for lev_base
ax.hist(lev_base, bins=bin_edges, color='skyblue', alpha=0.9, label='lev_base', align='left')
# Plot histogram for lev_large
ax.hist(lev_large, bins=bin_edges, color='orange', alpha=0.5, label='lev_large', align='left')
# Plot histogram for lev_xl
ax.hist(lev_xl, bins=bin_edges, color='green', alpha=0.5, label='lev_xl', align='left')

# Set labels and title
ax.set_xlabel('Levenshtein distance')
ax.set_ylabel('Frequency')
ax.set_title('Distribution of lev_base, lev_large and lev_xl')

# Add legend
ax.legend()

# Show plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Count occurrences of 0 distances for each model size
zero_counts = [
    (df['lev_small'] == 0).sum(),
    (df['lev_base'] == 0).sum(),
    (df['lev_large'] == 0).sum(),
    (df['lev_xl'] == 0).sum(),
    (df['lev_xxl'] == 0).sum()
]

# Define model sizes
model_sizes = ['lev_small', 'lev_base', 'lev_large', 'lev_xl', 'lev_xxl']

# Create a bar plot
plt.figure(figsize=(10, 6))
plt.bar(model_sizes, zero_counts, color=['pink', 'skyblue', 'orange', 'green', 'purple'])
plt.xlabel('Model Size')
plt.ylabel('Count of 0 Distances')
plt.title('Count of 0 Distances for Different Model Sizes')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Calculate total number of samples
total_samples = len(df)

# Calculate percentages of 0 distances for each model size
zero_percentages = [
    ((df['lev_small'] == 0).sum() / total_samples) * 100,
    ((df['lev_base'] == 0).sum() / total_samples) * 100,
    ((df['lev_large'] == 0).sum() / total_samples) * 100,
    ((df['lev_xl'] == 0).sum() / total_samples) * 100,
    ((df['lev_xxl'] == 0).sum() / total_samples) * 100
]

# Define model sizes
model_sizes = ['lev_small', 'lev_base', 'lev_large', 'lev_xl', 'lev_xxl']

# Create a bar plot
plt.figure(figsize=(10, 6))
plt.bar(model_sizes, zero_percentages, color=['pink', 'skyblue', 'orange', 'green', 'purple'])
plt.xlabel('Model Size')
plt.ylabel('Percentage (%) of 0 Distances')
plt.title('Percentage (%) of 0 Distances for Different Model Sizes')
plt.ylim(0, 7)  # Set y-axis limit to 0-100%
plt.show()
```

```python
# Extract data for lev_base and lev_large
lev_base = df['lev_base']
lev_large = df['lev_large']

# Sort the data
lev_base_sorted = np.sort(lev_base)
lev_large_sorted = np.sort(lev_large)

# Calculate the cumulative distribution function
cdf_base = np.arange(1, len(lev_base_sorted) + 1) / len(lev_base_sorted)
cdf_large = np.arange(1, len(lev_large_sorted) + 1) / len(lev_large_sorted)

# Create a figure and axis objects
plt.figure(figsize=(10, 6))

# Plot CDF plot for lev_base
plt.plot(lev_base_sorted, cdf_base, color='skyblue', label='lev_base')
# Plot CDF plot for lev_large
plt.plot(lev_large_sorted, cdf_large, color='orange', label='lev_large')

# Set labels and title
plt.xlabel('Levenshtein Distance')
plt.ylabel('Cumulative Probability')
plt.title('Cumulative Distribution Function of Levenshtein Distances')
plt.grid(True)

# Add legend
plt.legend()

# Show plot
plt.show()
```
